```python
l = ["apples", "bananas", "tomatoes"]
for item in l:
    if item == "bananas":
        break # Stop this loop right here

print(item)
```

```python
l = ["apples", "bananas", "tomatoes"]
for item in l:
    if item == "bananas":
        continue # Skip this particular iteration of the loop

print(item)
```
